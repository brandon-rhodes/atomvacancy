# Finding Gaps in a Lattice

The problem has been brought to my attention of automating the task of finding gaps in a hexagonal lattice of atoms, in a way that will not generate false positives as Brownian motion jostles the atoms.

I suspect that the right approach is probably a brute-force search across an $x\times y$ grid of points that are close enough together that any interesting gap is guaranteed to “set off” a detection function evaluated at each point on the grid. The detection function that seems natural is the distance to the nearest atomic center — or perhaps the square of the distance, to avoid computing extra square roots?

Let us open a sample file of “movie frames” that each has a moment-in-time list of atomic positions, and read in the first “frame” full of positions:

```python
lines = iter(open('dump.vacancy'))
for line in lines:
    if line.startswith('ITEM: ATOMS'):
        ax = []
        ay = []
        for line in lines:
            if line.startswith('ITEM: '):
                break
            fields = line.split()
            ax.append(float(fields[0]))
            ay.append(float(fields[1]))
        ax = array(ax)
        ay = array(ay)
        break
```

The above input routine results in two arrays, `ax` with the $x$-coordinates and `ay` with the $y$-coordinates.

We can view the atomic matrix at this moment in time by simply plotting these positions:

Given the gap-detection function that I proposed above, what does the “square of the distance to the nearest atom” function look like if rendered as a field across the area depicted above? Using a few NumPy vector operations, we can evaluate this function across the above area and then display it:

```python
xspace = linspace(9.0, 11.5, 100)
yspace = linspace(8.0, 11.0, 100)

X, Y = meshgrid(xspace, yspace)
X = X.flatten()
Y = Y.flatten()

xsquared = ((X[:,newaxis] - ax[newaxis,:]) ** 2.0)
ysquared = ((Y[:,newaxis] - ay[newaxis,:]) ** 2.0)

z = (xsquared + ysquared).min(axis=1)
z = z.reshape((100, 100))
```

```python
contourf(xspace, yspace, z, 20, cmap=cm.Oranges)
axes().set_aspect('equal', 'datalim')
```

If we want to find the center of the above gap in an automated fashion — always imagining that, in the future, the algorithm may be applied to much larger arrays from bigger simulations — then there are at least two approaches.

The brute-force approach, as noted earlier, would be to evaluate our gap-finding function at very many points that are close enough together that one of them is bound to land somewhere in the “deep” part of the gap where the gap-finding function reaches a value larger than we would expect it to gain between any three normally-deployed atoms. Having found some interesting points, we could then use a solver to find the exact point at the center of the gap.

Alternately, we could choose a larger grid of test points and, from each of them, start a search for the bottom of the nearest well. I suspect that this will be slower: minimization finders are not terribly fast beasts, and setting hundreds of them running across each frame of an animation does not sound efficient. But, I suppose, that depends on how many extra points would have to be thrown in to guarantee that one lands somewhere near the middle of each gap — adding more and more points to an $x\times y$ grid can also be expensive. But it seems like it could be vectorized more easily, whereas launching optimization steps with `scipy` has to be done one at a time.

Let me quickly illustrate running optimization searches. It looks like there is a sizeable region, at least $1.0\times 1.0$ in size, around the gap in which our function is at every point downhill towards the gap itself. Any search started in this region should converge on the gap itself:

```python
contourf(xspace, yspace, z, 20, cmap=cm.Oranges)
plot([10.0, 11.0, 11.0, 10.0, 10.0],
     [10.0, 10.0, 9.0, 9.0, 10.0])
axes().set_aspect('equal', 'datalim')
```

This means that, if we are going to start a search from each of a grid of points, they should not have to be any closer together than $1.0\times 1.0$ to be sure of finding any gaps in the atomic lattice.

So all we need to do is choose a `tx` and a `ty` that are $x$ and $y$ coordinates separated by $1.0$ each, visit each corner, and start up an optimizer at that location. Because `scipy` optimizers all seek minimization, our function here actually returns the negative of the square of the distance to the nearest atom:

```python
from scipy import optimize

contourf(xspace, yspace, z, 20, cmap=cm.Oranges)

def atom2(xy):
    "Return minus the square of the distance to the closest atom."
    x, y = xy
    dx = ax - x
    dy = ay - y
    return -(dx * dx + dy * dy).min()

tx = arange(9.0, 11.5, 1.0)
ty = arange(8.0, 11.0, 1.0)

for x in tx:
    for y in ty:
        xy = array([x, y])
        result = optimize.minimize(atom2, xy, method='Nelder-Mead', tol=0.0001)
        if result.success:
            print result.x, result.fun
            rx, ry = result.x
            color = 'ro' if result.fun < -0.5 else 'wo'
            plot(rx, ry, color)
    
axes().set_aspect('equal', 'datalim')
```

And we can see that the searches, started at the integer coordinate intersections of this grid, have indeed converged to the gaps between atoms. Most of them have a value somewhere around $-0.333$ indicating a normal lattice gap, but two of the searches found the “big gap” point whose value is around $-0.999$. By coloring these more extreme extrema red, we have at least a simple automated way of telling gaps where an atom is missing apart from run-of-the-mill gaps created by the lattice.
